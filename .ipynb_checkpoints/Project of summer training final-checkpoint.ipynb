## Remove Outlier at each column using IQR 

```python
#Remove Outlier of price column
for i in range(1,550):
    Q1=df['meas_length'].quantile(0.25)
    Q3=df['meas_length'].quantile(0.75)
    IQR=Q3-Q1
    lower_bound=Q1-1.5*IQR
    upper_bound=Q3+1.5*IQR
    filtered_df=df[(df['meas_length']>=lower_bound) & (df['meas_length']<=upper_bound)]
df=filtered_df
```

## Model Training

```python
# Correcting the selection of independent variables
x = df.loc[:,['color','clarity','carat_weight','cut_quality','lab','symmetry','polish','culet_size','culet_condition','depth_percent',
              'table_percent','meas_length','meas_width','meas_depth','girdle_min','girdle_max','fluor_color']]
# Correcting the selection of the dependent variable
y = df.loc[:,'cut']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
# training set 
import numpy as np
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Feature scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
#model building 
from sklearn.linear_model import LogisticRegression
```

```python
classifier = LogisticRegression()
```

```python
classifier.fit(x_train,y_train)
```

```python
y_pred = classifier.predict(x_test)
y_pred
```

## Accuracy

```python
#accuracy score
from sklearn.metrics import accuracy_score,classification_report
```

```python
score1 = accuracy_score(y_pred,y_test)
print('Accuracy using Logistic Regression: ',score1)
```

```python
report = classification_report(y_test, y_pred)
print(report)
```

```python
from sklearn.metrics import confusion_matrix
# Generate the confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Print the confusion matrix
print("Confusion Matrix:\n", cm)
```
